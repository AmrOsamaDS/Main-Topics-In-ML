# *حددت اسماء الاعمدة اللي الترتيب فيها ليه معنا والعس*

```python
ordinal_cols = ['power', 'price_status']
nominal_cols = ['type', 'color', 'physical_type']
```

# استخدمت الكولمن ترنسفورمر عشان يسهل عليا الشغل

# تطبيق عمليات معالجة مختلفة على أعمدة مختلفة في نفس الوقت

# قبل اي حاجة حددت ترتيب القيم داخل كل عمود يهموا الترتيب

```python
power_order = ['low', 'medium', 'high']
price_status_order = ['down', 'up']
```

## استدعيت الكولمن ترنسفورمر وهطبقوا عشان يعالج اعمدة مختلفة في نفس الوقت

```python
from sklearn.compose import ColumnTransformer
#  خلي بالك عملت ترتيب للقيم بتاعتي 
preprocessor = ColumnTransformer(
    transformers=[
        ('nom', OneHotEncoder(), nominal_cols),
        ('ord', OrdinalEncoder(categories=[power_order, price_status_order]), ordinal_cols)
    ])

X_encoded = preprocessor.fit_transform(df)
```

### الناتج اللي طلع هو مصفوفةبدون اسماء اعمدة ، فحولتها لداتا فريم ورجعت اسماء الاعمدة

```python
X_encoded=pd.DataFrame(X_encoded, columns=preprocessor.get_feature_names_out())
```
